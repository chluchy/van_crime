# Code for cleaning crime (from kaggle) and merging with demo (manipulated in excel by CH

```python
crime = pd.read_csv('crime_raw.csv')
demo = pd.read_csv('2016_demographics.csv')
crime.NEIGHBOURHOOD.unique()
```

### Create Date Column

```python
crime['DATE'] = pd.to_datetime(crime[['YEAR', 'MONTH', 'DAY']])
```

```python
crime.head()
```

### Clean neighbourhood so match the shape file.  Remove two neighbourhoods that don't exist in shape file.

```python
crime['NEIGHBOURHOOD']=crime['NEIGHBOURHOOD'].replace('Arbutus Ridge','Arbutus-Ridge')
crime['NEIGHBOURHOOD']=crime['NEIGHBOURHOOD'].replace('Central Business District','Downtown')
crime=crime[(crime.NEIGHBOURHOOD != 'Stanley Park') & (crime.NEIGHBOURHOOD != 'Musqueam') ]
crime.NEIGHBOURHOOD.unique()
```

### Drop rows that have null in Neighbourhood

```python
crime.dropna(subset=['NEIGHBOURHOOD'], inplace=True)
crime.reset_index()
len(crime)
```

```python
crime.to_csv('crime_clean.csv')
```

### Create new df just for 2016 and merge demographics data

```python
crime_sub = crime[crime.YEAR == 2016]
crime_2016 = pd.merge(crime_sub,demo, left_on= 'NEIGHBOURHOOD', right_on= 'neighbourhood',how='left')
crime_2016.to_csv('crime2016.csv')
```
